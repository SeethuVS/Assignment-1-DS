5. v) Scatter plot of SL v/s SW and SL v/s PW in a single plot

```python
plt.figure(figsize=(8,6))

plt.subplot(2,1,1)
plt.scatter(data['SL'],data['SW'],c='red')
plt.title('Plot of SL v/s SW')
plt.xlabel('SL')
plt.ylabel('SW')

plt.subplot(2,1,2)
plt.scatter(data['SL'],data['PW'],c='blue')
plt.title('Plot of SL v/s PW')
plt.xlabel('SL')
plt.ylabel('PW')

plt.tight_layout()
```
